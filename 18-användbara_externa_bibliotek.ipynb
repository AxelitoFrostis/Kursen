# Installera moduler

Installationen görs med komandot pip install följ av modulens namn, i terminalen. Mer info om det på deras [hemsida](https://pypi.org/). Ett exempel kan se ut som föjande:
```bat
pip install requests
```
Om inte det fungerar testa på windows
```bat
python -m pip install requests
```
Mac och linux
```sh
python3 -m pip install requests
```

Ett annat knep när man installerar moduler till projekt är att sammla dem i en text fil. Moduler ett projekt kräver kallas requirements eller dependencies, kort deps. I python brukar man därför samla alla i en map som heter just requirements.txt. I andra språk är andra namn vanliga. En sån fil kan se ut
```
requests
matplotlib
```
I terminalen, se till att du har requirements.txt i samma mapp som du arbetar i just nu, kolla annars det med dir eller ls beronde på operativsystem. Installera sedan med:
```sh
pip install -r requirements.txt
```
Då installeras alla deps automatiskt. 

En annan sak som kan vara bra att veta är att det är dumt att installera alla dependencies globalt, vilket man gör om man använder kommandot på det sättet. Istället skapa en virtuell miljö, kallat enviroment eller virtual enviroment (kort env eller venv). Detta gör med små moduler som man måste ha globalt. Det finns en inbyggd i python som heter just venv, men det finns en förbättrad version som man måste installera. Gör det:
```sh
python3 -m pip install --user virtualenv 
```
Exkludera trean om du använder windows. 

Kolla länken till virtualenv och lär dig använda det! Det är alltid bättre även om det tar mer plats på datorn. Det är både säkre, tydligare och minimerar buggar. [länk](https://virtualenv.pypa.io/en/latest/)

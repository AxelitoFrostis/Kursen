# Fler Inbyggda Funktioner

Hittills har vi bara tagit upp några få funktioner, print() och input(). Nu ska vi gå igenom flera användbara samt inbyggda bibliotek som finns tillsammans med pythoninstallationen. 

## Inbyggda funktioner

### abs(int)

Ger tillbaka det absoluta värdet på ett tal, ex abs(5) = 5, abs(-3) = 3. 
Kort och gott tar det bort -. Mycket användbart. 

### all (list)

Tar en lista (eller annat objekt med flera värden som vi inte gått igenom än) och ger tillbaka True om alla värden är sanna eller om listan är tom.

### ascii (str)

Skriver om en text till bara ascii. Vissa databaser, filer och andra datasystem kan inte hantera alla tecken i unicode (ex. å,ä, & ö) och skriver därför om dem. 

### bin (int)

Skriver om en siffra till ett binärt tal. 

### callable (object)

Kollar om det går att köra objektet. Svarar True eller False. Ex
```
callable (print) --> True
callable(3) --> False
```
har aldrig använt. 

### dir(object)

Svarar vilken information som är tillgänglig om objektet. Alla metoder och attribut. Svaret kommer i form av en lista. 

### enumerate(list)

Numrerar en lista på olika sätt. ett exempel:
```
enumerate(['äpple', 'päron', 'banan']) --> [(0, 'äpple'), (1, 'päron'), (2, 'banan')]
```

### filter(function, list)

Filtrerar en lista med hjälp av en funktion. Funktionen måste svara antingen True eller False. Den går egentligen igenom alla värden och kör funktionen på dem. Om funktionen ger true får de stanna annars tas de bort från svaret. 

### format()

Formaterar text, komplext hur det funkar och finns ofta bättre lösningar. [Här är en källa om hur man använder funktionen](https://docs.python.org/3/library/string.html#formatspec)

### globals()

Ger tillbaka den dictionary där alla dina globala variabler finns sparade. Det gör att du programatiskt kan ändra variablerna i ditt globala scope. Ofta är det bättre att använda andra dictionaries men ibland är den användbar. 

### hash()

Omvandlar ett objekt till en lång siffra som är unik för just ditt input. Det sjuka är att det ändå knappt finns något mönster. Det ända sättet att lista ut vad man stoppade in för värde är att testa sig fram. Det är det här som gör bitcoin säkert. Det är däremot väldigt avancerat, det får du kolla upp på egen hand. 

### hex(int)

Omvandlar ett tal till ett hexadecimalt tal (base16). 

### id(object)

Ger tillbaka objektets id. Det är i sjävla verket objektets minnesadress, mer om det när du lär dig C :)

### len()

Ger tillbaka objektets längd. Funkar på listor och texter. 

### locals()

Funkar på samma sätt som globals fast för variablerna i ditt lokala scope. 

### map(function, list)

Förändrar alla värden i listan med funktionen du gav. 

### max(list)

Ger tillbaka det största värdet i listan

### min(list)

Ger tillbaka det minsta värdet i listan

### open()

Används för att läsa filer på filsystemet. Mycket användbar. Ganska komplex i hur den används så bifogar därför en [länk](https://docs.python.org/3/library/functions.html#open). Bra läsning!

### repr(object)

Skriver om objektet till text. 

### reversed(list)

Ger tillbaka listan fast baklänges. 

### round(float)

Avrundar ett decimaltal till en int. Det går att ställa in hur mycket man ska avrunda. 

### sorted(list)

Ger tillbaka listan sorterad

### sum(list)

Summerar hela listan

### zip(list, list)

Tar två listor och parar ihop värdena med varandra efter index.
